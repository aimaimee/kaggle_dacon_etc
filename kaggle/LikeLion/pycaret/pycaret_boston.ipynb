```python
from pycaret.datasets import get_data

data = get_data('boston')
data.info()
```

```python
# 데이터 자동 전처리 및 파이프라인 생성
# 보스턴 데이터는 회귀
# setup은 automl 할 데이터, 종속변수(target), 학습/테스트 데이터 분리, kfold검증, 스케일링(default=z-score), 아웃라이어 제거, one-hot-encoding 등 가능
# https://pycaret.readthedocs.io/en/latest/api/regression.html 셋업 옵션
# setup은 파이프라인

from pycaret.regression import *

reg_test_1 = setup(data=data,
                   target = 'medv',
                   train_size = 0.8,
                   fold=5)
```

```python
# 알고리즘 자동 비교
# 나는 왜 노란색 안뜨지

best = compare_models(sort='mse')
```

```python
best
```

```python
# 모델 튜닝
best_tune = tune_model(best)
```

```python
best_tune
```

```python
# 모델 학습 정보 시각화

evaluate_model(best_tune)
```

```python
# 예측값 보기
predict_model(best_tune)
```
